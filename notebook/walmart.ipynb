```python
# import libraries

import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/Users/abdulrahim/Projects/sql + python/zero_analyst/Walmart.csv', encoding_errors = 'ignore' )
```

```python
#make column names lower case and replace space with underscore

df.columns = df.columns.str.lower()
```

```python
# convert date column to date data type

df['date'] = pd.to_datetime(df['date'], format='%d/%m/%y')
```

```python
df.info()
```

```python
df.head()
```

```python
# remove $ sign from unit_price and change data type to float

df['unit_price'] = df['unit_price'].str.replace("$","")

df['unit_price'] = df['unit_price'].astype(float)
```

```python
df.head()
```

```python
df.info()
```

```python
# checking duplicates

df.duplicated().sum()
```

```python
# removing duplicates

df.drop_duplicates(inplace=True)

df.duplicated().sum()
```

```python
df.shape
```

```python
# checking missing values

df.isnull().sum()
```

```python
# dropping all rows with missing records

df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df.columns
```

```python
# adding a new "total" column

df["total"] = df['unit_price'] * df['quantity']

df.head()
```

```python
df.columns
```

```python
# Connect to PostgreSQL Using SQLAlchemy

from sqlalchemy import create_engine

engine = create_engine(
    "postgresql+psycopg2://postgres:8801010734@localhost:5432/walmart_db"
)

print("Connected Successfully!")
```

```python
# Load DataFrame into PostgreSQL Without Index
df.to_sql(
    name="walmart_sales",
    con=engine,
    if_exists="append",
    index=False
)

print("Data Loaded Successfully!")
```
